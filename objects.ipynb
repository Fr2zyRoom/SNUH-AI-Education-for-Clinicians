```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./train.csv', header=None)
df.head()
```

```python
info = df[df[0] == './labels/4.jpg']
info.head()
```

```python
img = cv2.imread('./labels/4.jpg')
plt.imshow(img)
```

```python
for i in info.values:
    img = cv2.rectangle(img.astype(np.float32), (i[1], i[2]), (i[3], i[4]), (255, 0, 0), 5)
```

```python
plt.imshow(img.astype(np.uint8))
```

```python
h, w, _ = img.shape

n = 3
for i in np.arange(0, h, h//n):
    for j in np.arange(0, w, w//n):
        plt.imshow(img.astype(np.uint8)[i:i+h//2,j:j+w//2])
        plt.show()
```
